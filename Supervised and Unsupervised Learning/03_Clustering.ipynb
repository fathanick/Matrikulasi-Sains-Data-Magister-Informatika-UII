```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Number of data points
num_samples = 300

# Random seed for reproducibility
np.random.seed(42)

# Generate data for 3 clusters
data = np.vstack([
    np.random.randn(num_samples // 3, 2) + [2, 2],
    np.random.randn(num_samples // 3, 2) + [6, 6],
    np.random.randn(num_samples // 3, 2) + [10, 2],
])
```

```python
data
```

### K-Means Clustering

```python
# Number of clusters
k = 3

# Create a KMeans instance with the specified number of clusters
kmeans = KMeans(n_clusters=k)

# Fit the KMeans model to the data
kmeans.fit(data)

# Get the cluster labels
cluster_labels = kmeans.labels_

# Get the cluster centers
cluster_centers = kmeans.cluster_centers_
```

```python
# Define colors for the clusters
colors = ['red', 'green', 'blue']

# Plot the data points with different colors for each cluster
for i in range(k):
    plt.scatter(data[cluster_labels == i, 0], data[cluster_labels == i, 1], color=colors[i], label=f'Cluster {i+1}')

# Plot the cluster centers
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], color='black', marker='x', s=100, label='Centers')

# Add legend and show the plot
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```

### Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Number of clusters
n_clusters = 3

# Create the AgglomerativeClustering model
agglomerative = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')

# Fit the model to the data
cluster_labels = agglomerative.fit_predict(data)
```

```python
# Generate the linkage matrix for the dendrogram
linkage_matrix = linkage(data, method='ward')

# Plot the dendrogram
plt.figure()
dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
